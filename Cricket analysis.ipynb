```python
mean_value = data["Strike_Rate"].mean()
print(mean_value)

means_value = data["Average"].mean()
print(means_value)

means_values = data["Runs"].mean()
print(means_values)
```

```python
Capped_Allrounders["Strike_Rate"].fillna(value=mean_value, inplace=True)
Capped_Allrounders["Average"].fillna(value=means_values, inplace=True)
Capped_Allrounders["Runs"].fillna(value=means_value, inplace=True)


print(Capped_batters.isna().sum())
print(Capped_Allrounders.isna().sum())

#Performing this let there is no null values available in my data  a part of data cleaning for better result
```
